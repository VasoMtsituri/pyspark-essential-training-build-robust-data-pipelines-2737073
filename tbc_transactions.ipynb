# Import and initialize PySpark session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("TBC_TRANSACTIONS") \
    .getOrCreate()
```

```python
file_path = "/content/drive/MyDrive/Statement.csv"

rdd = spark.sparkContext.textFile(file_path)

# While TBC statement files have two headers (Georgian and English), skip first and make second as real header
df = rdd.zipWithIndex().filter(lambda x: x[1] > 0).map(lambda x: x[0])

transactions_raw = spark.read.csv(df, header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import col

transactions_raw.select(col("`Op. Code`")).distinct()
```

```python
transactions_raw.filter(col("`Op. Code`") == 'GMB')
```
